```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
amount_value_array=df['Amount'].values
```

```python
amount_value_array
```

```python
df['Amount']=sc.fit_transform(amount_value_array.reshape(-1,1))
```

```python
df['Amount']
```

```python
x=df.drop('Class',axis=1)
```

```python
x
```

```python
y=df['Class']
```

```python
y
```

```python
#data split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#decion tree
```

```python
from sklearn.tree import DecisionTreeClassifier
model_tree=DecisionTreeClassifier(max_depth=4,criterion='entropy')
```

```python
model_tree.fit(x_train,y_train)
```

```python
y_pred_T=model_tree.predict(x_test)
```

```python
#logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_logistic=LogisticRegression()
```

```python
model_logistic.fit(x_train,y_train)
```

```python
y_pred_log=model_logistic.predict(x_test)
```

```python
#random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_forest=RandomForestClassifier(max_depth=4)
```

```python
model_forest.fit(x_train,y_train)
```

```python
y_pred_forest=model_forest.predict(x_test)
```

```python
#model evaluation
```

```python
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix,f1_score
```

```python
#1.Accuracy score
```

```python
accuracy_tree=accuracy_score(y_test,y_pred_T)
```

```python
accuracy_log=accuracy_score(y_test,y_pred_log)
```

```python
accuracy_forest=accuracy_score(y_test,y_pred_forest)
```

```python
print(f'accuracy by decision tree model is {accuracy_tree} ')
print(f'accuracy by logistic regression model is {accuracy_log} ')
print(f'accuracy by random forest  model is {accuracy_forest} ')
```

```python
#2.precision score
```

```python
precion_tree=precision_score(y_test,y_pred_T,average=None)
```

```python
precion_log=precision_score(y_test,y_pred_log,average=None)
```

```python
precion_forest=precision_score(y_test,y_pred_forest,average=None)
```

```python
print(f'precision by decision tree model is {precion_tree} ')
print(f'precision by logistic regression model is {precion_log} ')
print(f'precision by random forest  model is {precion_forest} ')
```

```python
#3. F1 score
```

```python
f1_score_tree=f1_score(y_test,y_pred_T,average=None)
```

```python
f1_score_log=f1_score(y_test,y_pred_log,average=None)
```

```python
f1_score_forest=f1_score(y_test,y_pred_forest,average=None)
```

```python
print(f'f1 score by decision tree model is {f1_score_tree} ')
print(f'f1 score by logistic regression model is {f1_score_log} ')
print(f'f1 score by random forest  model is {f1_score_forest} ')
```
